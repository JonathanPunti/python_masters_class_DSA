### Task 7
Which year has the most failed banks?

```python
banklist['Closing Date'] = pd.to_datetime(banklist['Closing Date'])
banklist['Year'] = banklist['Closing Date'].dt.year
task7 = banklist['Year'].value_counts().idxmax()
task7
```

### Task 8

```python
banklist['Month'] = banklist['Closing Date'].dt.month_name()
task8 = banklist['Month'].value_counts().idxmax()
task8
```
