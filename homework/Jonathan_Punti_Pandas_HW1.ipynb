### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
def calculate_grade(score): # Define a function  returning string
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

grad_Mates['CS5000 Grade'] = grad_Mates['CS 5000'].apply(calculate_grade) # apply function to scores
grad_Mates
```

### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2023, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
def score_to_gpa(score): # function to return gpa integer
    if score >= 90:
        return 4.0
    elif score >= 80:
        return 3.0
    elif score >= 70:
        return 2.0
    elif score >= 60:
        return 1.0
    else:
        return 0.0

gpa_cs5000 = grad_Mates['CS 5000'].apply(score_to_gpa)
gpa_cs5110 = grad_Mates['CS 5110'].apply(score_to_gpa)
gpa_cs5600 = grad_Mates['CS 5600'].apply(score_to_gpa)

grad_Mates['GPA'] = (gpa_cs5000 * 3 + gpa_cs5110 * 3 + gpa_cs5600 * 3) / 9 # reviewed the code to calculate average and put ouput into a new column
grad_Mates
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
+ `Normal: 3.0 <= GPA < 3.5`
+ `In Probation: GPA < 3.0`

```python
def determine_status(gpa): # define function with integer as input and returning a string
    if gpa >= 3.5:
        return 'Honor Roll'
    elif gpa >= 3.0:
        return 'Normal'
    else:
        return 'In Probation'

# Apply the determine_status

grad_Mates['Status'] = grad_Mates['GPA'].apply(determine_status)
grad_Mates
```
