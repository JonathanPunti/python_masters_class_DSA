### Task 7
Which year has the most failed banks?

```python
banklist['Closing Date'] = pd.to_datetime(banklist['Closing Date'])
banklist['Year'] = banklist['Closing Date'].dt.year
most_failed_banks_year = banklist['Year'].value_counts().idxmax()
most_failed_banks_year
```

### Task 8
Which month has the most failed banks?

```python
banklist['Month'] = banklist['Closing Date'].dt.month_name()
most_failed_banks_month = banklist['Month'].value_counts().idxmax()
most_failed_banks_month
```

### Task 9
